```python
import os
import csv

# Define gene_list and file paths
gene_list = 250
input_file_path = 'documents.txt'
output_file_path = f'./Data/Reference/persona_{gene_list}_genes.csv'

# Ensure the output directory exists
os.makedirs(os.path.dirname(output_file_path), exist_ok=True)

# Read the input file with UTF-8 encoding
with open(input_file_path, 'r', encoding="utf-8") as file:
    data = file.read()

# Split the text into references based on the delimiter '|'
references = data.split('|')
# Clean references and remove empty lines
references = [ref.strip() for ref in references if ref.strip()]

# Convert to absolute path to avoid confusion about relative directories
absolute_output_path = os.path.abspath(output_file_path)

# Write to CSV using the csv.writer
with open(absolute_output_path, 'w', newline='', encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    for ref in references:
        writer.writerow([ref])
        print(f"Writing reference: {ref}")  # Debug print to console

print(f"CSV file has been created and saved to: {absolute_output_path}")
```
