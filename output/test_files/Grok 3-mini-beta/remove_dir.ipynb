```python
#!/usr/bin/env python3
import os
import re

# match grok-3-mini-fast-beta-GSEA-<3digits>-<1digit>-<anyDigits>.<anyDigits>.txt
pattern = re.compile(
    r"^grok-3-mini-fast-beta-GSEA-(\d{3})-(\d)-(\d+\.\d+)\.txt$"
)

def main():
    for fname in os.listdir("."):
        m = pattern.match(fname)
        if not m:
            continue

        # we only need the 1-digit and the full decimal
        _, single_digit, decimal = m.groups()
        new_name = f"grok-3-mini-beta-GSEA-{single_digit}-{decimal}.txt"

        print(f"Renaming:\n  {fname}\n→ {new_name}\n")
        os.rename(fname, new_name)

if __name__ == "__main__":
    main()
```
